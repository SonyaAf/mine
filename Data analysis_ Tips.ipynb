```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
tips = sns.load_dataset('tips') 

tips.describe()
```

```python
tips.isna()
```

```python
tips.isnull().sum()
```

```python
print(tips.shape)
```

```python
tips.head()
 
```

```python
tips.plot(y='total_bill', kind='hist', bins=30, range=(4,8), cumulative=True, normed=True,color='black')
plt.title('Cumulative distribution function (CDF)') 
plt.show() 
```

```python
ax = sns.boxplot(x="day", y="total_bill", data=tips,palette='cubehelix')
```

```python
ax = sns.boxplot(x="day", y="tip", data=tips,palette='cubehelix')
```

```python
ax = sns.boxplot(x="sex", y="size", data=tips,palette='cubehelix')
```

```python
sns.catplot(x="sex", y="total_bill", hue="time", kind="bar", data=tips)
```

```python
sns.catplot(x="sex", y="total_bill", hue="day", kind="bar", data=tips)
```

```python
sns.catplot(x="sex", y="total_bill", hue="smoker", kind="bar", data=tips)
```

```python
sns.catplot(x="smoker", y="total_bill", hue="time", kind="bar", data=tips)
```

```python
sns.pairplot(tips)
```

```python
sns.distplot(tips['total_bill']).set_title('Distribution total bill') 
```

```python
sns.distplot(tips['tip']).set_title('Distribution tips')
```

```python
sns.distplot(tips['size']).set_title('Distribution size')
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex') 
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',palette='Set1')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker',palette='husl')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day',palette='Blues_d') 
plt.show()
```

```python
sns.residplot(x='total_bill',y='tip',data=tips,color='indianred').set_title('Residuals') 
plt.show()
```

```python
corr=tips.corr()
print(corr)
```

```python
import seaborn as sns
f,ax=plt.subplots(figsize=(18,10))
sns.heatmap(corr,linewidths=2.0,ax=ax,annot=True)
ax.set_title('Correlation Matrix')
```
