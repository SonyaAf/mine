```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.DataFrame(np.random.randint(0,100,size=(50, 1)))
df2 = pd.DataFrame(np.random.randint(0,100,size=(50, 1)))
df3 = range (5,55)
```

```python
#Multiple plots on single axis

plt.plot(df3, df1, color='black')

plt.plot(df3, df2, color='red')
plt.title('random plot')
# Display the plot
plt.show()
```

```python
# Create plot axes [xlo, ylo, width, height] for the first line plot
plt.axes([0.05,0.05,0.425,0.9])

# Plot in black the % df1
plt.plot(df3, df1, color='black')

# Create plot axes [xlo, ylo, width, height] for the second line plot
plt.axes([0.525,0.05,0.425,0.9])

# Plot in red the % df2
plt.plot(df3, df2, color='red')

# Display the plot
plt.show()
```

```python
# Create a figure with 1x2 subplot and make the left subplot active
plt.subplot(1,2,1)

# Plot in black the % df1
plt.plot(df3, df1, color='black')
plt.title('df1')

# Make the right subplot active in the current 1x2 subplot grid
plt.subplot(1,2,2)

# Plot in red the % df2
plt.plot(df3, df2, color='red')
plt.title('df2')

# Use plt.tight_layout() to improve the spacing between subplots
plt.tight_layout()
plt.show()
```

```python
df4 = pd.DataFrame(np.random.randint(0,100,size=(50, 1)))
df5 = pd.DataFrame(np.random.randint(0,100,size=(50, 1)))
```

```python
# Create a figure with 2x2 subplot layout and make the top left subplot active
plt.subplot(2, 2, 1) 

# Plot in black the % df1
plt.plot(df3, df1, color='black')
plt.title('Physical Sciences')

# Make the top right subplot active in the current 2x2 subplot grid 
plt.subplot(2, 2, 2)

# Plot in red the % df2
plt.plot(df3, df2, color='red')
plt.title('df2')

# Make the bottom left subplot active in the current 2x2 subplot grid
plt.subplot(2, 2, 3)

# Plot in green the % df4
plt.plot(df3, df4, color='blue')
plt.title('df4')

# Make the bottom right subplot active in the current 2x2 subplot grid
plt.subplot(2, 2, 4)

# Plot in yellow the % df5
plt.plot(df3, df5, color='yellow')
plt.title('df5')

# Improve the spacing between subplots and display them
plt.tight_layout()
plt.show()

#save plot
plt.savefig('random_plot.png')
```

```python
import matplotlib.patches as mpatches
# Plot the % of df1 and df2
plt.plot(df3,df1, color='red',label='df1') 
plt.plot(df3, df2, color='blue',label='df2')

# Add the axis labels
plt.xlabel('x')
plt.ylabel('y')

# Add a legend at the upper right
plt.legend(loc='upper right') 

# Set the x-axis range
plt.xlim((0,65))

# Set the y-axis range
plt.ylim((0,150))

# Add a title and display the plot
plt.title('Random plot')
plt.show()

```

```python
# Set the style to 'ggplot'
plt.style.use('ggplot')

# Create a figure with 2x2 subplot layout
plt.subplot(2, 2, 1) 

# Plot the enrollment % df1
plt.plot(df3, df1, color='blue')
plt.title('df1')

# Plot the enrollment % df2
plt.subplot(2, 2, 2)
plt.plot(df3, df2, color='red')
plt.title('df2')

# Add annotation
cs_max = df2.max()
yr_max = df3.max()
plt.annotate('Maximum', xy=(yr_max, cs_max), xytext=(yr_max-1, cs_max-10), arrowprops=dict(facecolor='black'))

# Plot the enrollmment % df4
plt.subplot(2, 2, 3)
plt.plot(df3, df4, color='green')
plt.title('df4')

# Plot the enrollment % of df5
plt.subplot(2, 2, 4)
plt.plot(df3, df5, color='yellow')
plt.title('df5')

# Improve spacing between subplots and display them
plt.tight_layout()
plt.show()
```

```python
x = np.random.randn(100) 
y = np.random.randn(100)/10 

# Generate a 2-D histogram
plt.hist2d(x,y,bins=(15,15))

# Add a color bar to the histogram
plt.colorbar() 

# Add labels, title, and display the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('hist2d() plot')
plt.show()
```

```python
x = np.random.randn(100) 
y = np.random.randn(100)/10 

# Generate a 2d histogram with hexagonal bins
plt.hexbin(x,y,gridsize=(20,15))
           
# Add a color bar to the histogram
plt.colorbar() 

# Add labels, title, and display the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('hexbin() plot')
plt.show()
```
